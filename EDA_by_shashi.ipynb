```python
# Day 01 â€“ Basic EDA

from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
```

```python
# Display first rows
print("First 5 Rows:")
print(df.head())
```

```python
# Dataset info
print("\nDataset Info:")
df.info()
```

```python
# Statistical summary
print("\nDescription:")
print(df.describe())
```

```python

# Null values
print("\nNull Values:")
print(df.isnull().sum())
```

```python
# Histogram
df.hist(figsize=(10,8))
plt.show()
```
